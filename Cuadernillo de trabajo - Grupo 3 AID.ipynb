```python
# El dataset principal tiene que ver con la prevalencia de enfermedades mentales en el mundo, entre los años 1990 y 2019.
# Las enfermedades mentales analizadas son: Esquizofrenia, depresión, ansiedad, trastorno bipolar de la personalidad y desórdenes alimenticios.
df_MH_1.head()
```

```python
# El tercer dataset tiene que ver con el ingreso PIB per capita (GPD por las siglas en ingles), entre los años 1960 y 2020. Las medidas ya están estandarizadas.
# Las enfermedades mentales analizadas son: Esquizofrenia, depresión, ansiedad, trastorno bipolar de la personalidad y desórdenes alimenticios.
df_GPD.head()
```

```python
# El sexto dataset tiene que ver con el porcentaje de la población por cada 100 personas que son dependientes de drogas o alcohol, para cada género.
# Se define como dependencia la presencia de tres o más indicadores por al menos un mes en el último año.
substances_disorders_df.head()
```

```python
# El séptimo dataset tiene que ver con el porcentaje de población que tiene problemas con el alcohol, contrastados con el porcentaje de población que tiene problemas
# con las drogas.
alcohol_or_drugs_users_df.head()
```

```python
# En vista que utilizaremos muchos datasets en nuestro análisis, y ya se apreciaron valores NaN al utilizar el método ".head()", realizaremos un merge entre los datos,
# Esto con la finalidad de agilizar le filtrado y pre-procesamiento de datos, así como brindar una flujo de trabajo posterior más cómodo para cada hipótesis.
# Cada dataset presenta una columna país y código del mismo, lo que será fundamental al momento de unirlos.
```

```python
# Ptrimera unión de datos con el método "merge"
first_merge = pd.merge(df_MH_1, df_MH_2, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='inner')
first_merge.info()
```

```python
# Segunda unión de datos con el método "merge"
second_merge = pd.merge(first_merge, mental_illness_ratio_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='inner')
second_merge.info()
```

```python
# Tercera unión de datos con el método "merge"
third_merge = pd.merge(second_merge, substances_disorders_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='inner')
third_merge.info()
```

```python
# Cuarta unión de datos con el método "merge"
final_merge = pd.merge(third_merge, alcohol_or_drugs_users_df, left_on=['Entity', 'Code', 'Year'], right_on=['Entity', 'Code', 'Year'], how='inner')
final_merge.head()
```

```python
final_merge.info()
```

```python
# Finalmente nos quedamos con 3 datsets, 2 independientes: df_GDP y df_HDi. Y uno 1 creado a partir de merg: final_merge
# Evaluamos si hay valores nulos y nan, aunque el .info() nos muestra que si, pero queremos saber con exactitud donde.
final_merge.isnull().sum()
```

```python
# Nos percatamos que hay 3 columnas con valores nulos, por lo que se realiza el manejo correspondiente
# Eliminaremos las columna Population y Continent ya la ultima presenta valores nulos y no son de utilidad para el estudio.
final_merge.drop(['Population (historical estimates)_x', 'Population (historical estimates)_y', 'Continent_x', 'Continent_y'], axis=1, inplace=True)
final_merge.head()
```

```python
# Ya tenemos listo el dataset, por lo que se trabajaran con las hipotesis independientes
final_merge.rename(columns={'Entity': 'Country'}, inplace=True)
final_merge.info()
```

#Hipotesis 1


```python
# Hipotesis 1:
# “Los países con un Índice de Desarrollo Humano (IDH) más elevado presentan una menor prevalencia de trastornos mentales en comparación con aquellos países con un IDH más bajo.
# Se sugiere así una relación inversamente proporcional entre el Grupo de Desarrollo Humano de un país (dictado por el IDH) y la Prevalencia de Trastornos Mentales. Durante el año 2019”

df_hip_1 = final_merge.copy()

# Para realizar lo anterior nos quedaremos centraremos en datos del 2019, por lo que filtraremos para los datos de ese año.
# Una vez hecho esto también podremos acplar el dataset "df_HDI" con los datos correspondientes a ese año
df_hip_1 = df_hip_1[df_hip_1["Year"] == 2019]
df_hip_1.head()
```

```python
# Verificamos la anturaleza de los nombres de las columnas para realizar el filtrado correspondiente

df_HDI.head()
```

```python
# Notamos que tienen valores NaN, por lo que hay que limpiar. Pero antes de ello filtraremos para datos del 2019, y una vez hecho eso verificaremos si permanecen valors NaN

columns_2019 = df_HDI.filter(like='2019').dropna().columns
columns_2019
```

```python
# Rescatamos las columnas utiles que no presentan 2019 en el nombre y definimos el filtro

filter_HDI = list(df_HDI[["Entity", "Code", "Human Development Groups"]].columns) + list(columns_2019)
filter_HDI
```

```python
# Creamos un dataset con el filtro aplicado
df_HDI_filtred = df_HDI[filter_HDI]
df_HDI_filtred.head()
```

```python
# Nos quedaremos con las columnas que utilizaremos en el an[alisis]
df_HDI_filtred = df_HDI_filtred[["Entity", "Code", "Human Development Groups", "Human Development Index (2019)"]]
df_HDI_filtred.head()
```

```python
# Una vez filtrado lo m[as] posible el dataset procedemos a verificar la prevalencia de valores nulos
df_HDI_filtred.isnull().sum()
```

```python
# Hay valores nulos, pero al ser una porcion tan infima de la muestra los eliminaremos
df_HDI_filtred = df_HDI_filtred.dropna()
df_HDI_filtred.isnull().sum()
```

```python
# Una vez filtrado el dataset df_HDI uniremos su derivado al dataset final_merge, esto con la finalidad de agilizar la visualizacion de datos
df_HDI_filtred = df_HDI_filtred.rename(columns={"Entity":"Country"})
df_hip_1_merged = pd.merge(df_hip_1, df_HDI_filtred, on=["Country","Code"])
df_hip_1_merged.head()
```

```python
# Verificamos finalmente si hay nulos
df_hip_1_merged.isnull().sum()
```

```python
df_hip_1_merged.columns
```

```python
# Eliminaremos las columnas que no utilizaremos y quedaron depues del merge y cambiaremos el nombre de las residuales
cols_names = [
    'Country',
    'Code',
    'Year',
    'Schizophrenia_Share',
    'Depressive_Disorders_Share',
    'Anxiety_Disorders_Share',
    'Bipolar_Disorders_Share',
    'Eating_Disorders_Share',
    'Human_Development_Groups',
    'Human_Development_Index'
]

df_hip_1_merged = df_hip_1_merged[['Country', 'Code', 'Year',
       'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized',
       'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized',
       'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized',
       'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized',
       'Eating disorders (share of population) - Sex: Both - Age: Age-standardized',
       'Human Development Groups',
       'Human Development Index (2019)']]

df_hip_1_merged.columns = cols_names

# Aprovechamos de crear una columna que comprenda el promedio de tratornos mentales
df_hip_1_merged['Mental_Disorders_Prevalence'] = df_hip_1_merged[['Schizophrenia_Share', 'Depressive_Disorders_Share', 'Anxiety_Disorders_Share', 'Bipolar_Disorders_Share', 'Eating_Disorders_Share']].mean(axis=1)
df_hip_1_merged
```

```python
# Tenemos el dataset listo, por lo que antes de visualizar haremos un primer acercamiento estadistico
df_hip_1_merged.describe()
```

## Analisis Grafico

```python
# Preparar los grupos de desarrollo humano para graficar
df_hip_1_merged['Human_Development_Groups'] = pd.Categorical(df_hip_1_merged['Human_Development_Groups'], categories=['Low', 'Medium', 'High', 'Very High'], ordered=True)
```

```python
# Graficar HDI vs prevalencia de trastornos mentales por Grupo de Desarrollo HUmano - En este punto se da por hecho las divisiones de Grupo Humano
"""
Bajo (low) con un IDH inferior a 0.55
Medio (medium) con un IDH entre 0.55 y 0.70
Alto (high) con un IDH entre 0.70 y 0.80
Muy Alto (very high) con un IDH superior al 0.80 ap
"""

# Creamos el grafico de dispersion IDH vs Prevalencia de Desordenes Mentales
(ggplot(df_hip_1_merged, aes(x='Human_Development_Index', y='Mental_Disorders_Prevalence', color='Human_Development_Groups'))
+ geom_point()
+ labs(x='Indice Desarrollo Humano', y='% Prevalencia de Desordenes Mentales')
+ theme_linedraw()
+ theme(axis_text_x=element_text(rotation=90), figure_size=(8,6))
+ scale_color_manual(values=["#8cbf8f","#d9d18c","#8cb3d9","#d98c8c"])
+ labs(title='Grafico de Dispersion IDH v.s Prevalencia de Desordenes Mentales', x='Índice de Desarrollo Humano por Países', y='% Prevalencia de Desorden Mentales en la Poblacion', fill='Grupos de Desarrollo Humano')
 )
```

```python
# En esta sección, comparamos los países que tienen la mayor cantidad de trastornos mentales y la mayor tasa de consumo de sustancias
data_melted = filtered_df_hip_4.melt(id_vars=['Country', 'Year'],
                                 value_vars=['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized',
                                             'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized',
                                             'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized',
                                             'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized',
                                             'Eating disorders (share of population) - Sex: Both - Age: Age-standardized',
                                             'Current number of cases of substance use disorders per 100 people, in both sexes aged age-standardized'],
                                 var_name='Disorder', value_name='Cases')

# Seleccionamos los 10 países con mayores casos de trastornos mentales y consumo de sustancias
top_countries = data_melted.groupby('Country')['Cases'].sum().nlargest(10).index
data_top_countries = data_melted[data_melted['Country'].isin(top_countries)]

# Construimos un gráfico de barras apiladas para representar la prevalencia de consumo de sustancias y la prevalencia de trastornos mentales.
bar_plot = (ggplot(data_top_countries, aes(x='Country', y='Cases', fill='Disorder'))
            + geom_bar(stat='identity', position='stack')
            + facet_wrap('~Year')
            + ggtitle('Comparación de trastornos mentales y consumo de sustancias en los 10 países con mayores tasas (1999 vs 2019)')
            + labs(x='País', y='Porcentaje de casos por cada 100 personas', fill='Listado de trastornos')
            + theme(axis_text_x=element_text(hjust=0.5), plot_title=element_text(hjust=0.5), legend_position='bottom')
            + coord_flip())
bar_plot
```
